## Investigating the Relationship Between Speed and Crash Outcomes
Understanding the intricate relationship between vehicle speed and crash outcomes is paramount for enhancing road safety and mitigating the impact of traffic accidents. This investigation delves into three critical aspects of this relationship: the Correlation of Distance Affected by a Crash and an Accident's Speed at the Moment of the Accident; the Correlation of Total Number of Crashes and an Accident's Speed at the Moment of the Accident; and the Correlation of Accident Severity and an Accident's Speed at the Moment of the Accident. Through interactive visualizations and data analysis, this study aims to elucidate how variations in vehicle speed influence the severity, frequency, and spatial distribution of traffic accidents. By uncovering these insights, we can inform evidence-based interventions, policies, and engineering solutions to create safer road environments and reduce the societal and economic toll of traffic collisions.

```python
labels=["0-10 MPH", "10-20 MPH", "20-30 MPH", "30-40 MPH", "40-50 MPH", "50+ MPH"]
Query3SeverityPlot(raw_data, labels,"Query 3: Correlation of Accident Severity and an Accident's Speed at the Moment of the Accident")
```

The plot illustrating the Correlation of Accident Severity and an Accident's Speed at the Moment of the Accident provides a comprehensive visualization tool for examining how the speed of vehicles involved in accidents relates to the severity of those accidents. Users can interactively select time intervals to observe changes in accident severity trends across different speed categories, ranging from 0-10 MPH to 50+ MPH. This functionality facilitates a detailed exploration of how variations in vehicle speed impact the severity of accidents, potentially highlighting critical thresholds where higher speeds correlate with increased severity. 

The plot illustrating the Correlation of Accident Severity and an Accident's Speed at the Moment of the Accident reveals several  patterns. Despite variations, the lines representing different speed categories generally remain close to each other over time, suggesting a consistent relationship between speed and accident severity. Notably, there is a significant slump in accident severity for incidents involving speeds of 50+ MPH between January 2016 and March 2016, indicating a temporary reduction in severity during that period. Additionally, around 2018, there is a decrease in accident severity across all speed categories, coinciding with an increase in accidents involving speeds of 0-10 MPH during the same timeframe. These observations underscore the complex interplay between speed and accident severity, highlighting the need for further research to understand the underlying factors driving these trends. 

Researching the relationship between accident severity and an accident's speed is essential for several reasons. Firstly, it helps identify the role of speed as a contributing factor to accident severity, informing targeted interventions and policies aimed at reducing speed-related accidents and mitigating their consequences. Additionally, understanding this relationship aids in the development of evidence-based speed management strategies and road safety initiatives, contributing to the overarching goal of reducing traffic fatalities and injuries. Moreover, such research can inform the design and implementation of engineering solutions and traffic control measures to create safer road environments and promote responsible driving behaviors, ultimately enhancing overall road safety and saving lives.
