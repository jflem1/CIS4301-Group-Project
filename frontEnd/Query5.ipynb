## Investigating the Impact of Airport Proximity on Road Safety Outcomes
In this investigation, we delve into the intricate relationship between airport proximity and road safety outcomes through the analysis of three key correlations. By exploring the correlations between accident severity, number of crashes, distance affected by a crash, and the distance from the accident to an airport, we aim to uncover insights into how airport proximity influences road safety dynamics. As major transportation hubs, airports attract significant vehicular traffic, making understanding the spatial implications of their proximity crucial for informed decision-making in infrastructure planning and traffic management. By examining these correlations, we seek to identify patterns and trends that can inform targeted interventions, such as the installation of more viable roads and reasonable traffic lights, aimed at mitigating crash severity and improving overall road safety near airports.


```python
labels = ["Zero to Two Miles", "Two to Four Miles", "Four to Six Miles", "Six to Eight Miles", "Eight to Ten Miles", "Over Ten Miles"]
Query5SeverityPlot(query5_data, labels, "Query 5: Correlation of Accident Severity and Distance From the Accident to an Airport")
```

The plot illustrates the correlation between accident severity and the distance from the accident to an airport, offering users a dynamic tool to explore the spatial implications of accidents in proximity to airport facilities. With the option to drag and select specific time periods, users can analyze temporal variations in this relationship, facilitating the identification of trends and patterns over time. The plot features six distinct lines representing different distance ranges from the airport—ranging from 0-2 miles to over 10 miles—allowing users to compare accident severity across varying distances from airport facilities.

The plot reveals a consistent trend in accident severity across varying distances from the accident to an airport. Despite differences in proximity to the airport, the severity of accidents remains remarkably similar over time, with all six lines indicating severity levels ranging between 2.0 and 2.5. This trend suggests that the distance from the accident to the airport may not significantly influence the severity of accidents. Even as the distance increases from 0-2 miles to over 10 miles from the airport, there is little variation in severity levels, indicating a consistent pattern of accidents with moderate severity. This finding challenges the assumption that proximity to an airport might inherently increase the severity of accidents due to factors such as high traffic volume or increased speeds on highways leading to airports. Instead, it suggests that other factors, such as road conditions, driver behavior, or infrastructure design, may play a more prominent role in determining accident severity. Further analysis may be necessary to uncover the underlying factors driving this consistent trend and to inform targeted interventions aimed at reducing accident severity regardless of airport proximity.

Researching the relationship between accident severity and distance from the accident to an airport is crucial for several reasons. Firstly, airports typically serve as major transportation hubs, attracting significant vehicular traffic in surrounding areas. Understanding how accident severity correlates with proximity to airports can inform infrastructure planning and traffic management strategies aimed at reducing the likelihood and severity of accidents near these high-traffic zones. Additionally, given that higher speeds are often observed on highways leading to airports, where speed limits may be higher, investigating this relationship can shed light on the effectiveness of speed limit enforcement measures in mitigating crash severity. Ultimately, this research contributes to the development of evidence-based interventions to enhance road safety and reduce the societal costs associated with accidents near airports.
